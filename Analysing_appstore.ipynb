```python
df_clean[['App', 'Installs']].groupby('Installs').count()
```

```python
df_clean.Installs = df_clean.Installs.astype(str).str.replace(',', '')
df_clean.Installs = pd.to_numeric(df_clean.Installs)
df_clean[['App', 'Installs']].groupby('Installs').count()
```

```python
df_clean.Price.describe()
```

```python
df_clean.Price = df_clean.Price.astype(str).str.replace('$', '')
df_clean.Price = pd.to_numeric(df_clean.Price)
df_clean.sort_values('Price', ascending=False).head(20)
```

```python
df_clean = df_clean[df_clean['Price'] < 250]
df_clean.sort_values('Price', ascending=False).head()
```

```python
df_clean['Revenue_Estimate'] = df_clean.Installs.mul(df_clean.Price)
df_clean.sort_values('Revenue_Estimate', ascending=False).head()
```

```python
df_clean.Category.nunique()
```

```python
top_category = df_clean.Category.value_counts()[:10]
top_category
```

```python
bar = px.bar(x = top_category.index, y = top_category.values)
bar.show()
```

```python
category_installs = df_clean.groupby('Category').agg({'Installs': pd.Series.sum})
category_installs.sort_values('Installs', ascending=True, inplace=True)
```

```python
h_bar = px.bar(x = category_installs.Installs, y = category_installs.index, orientation='h', title='Category Popularity')
h_bar.update_layout(xaxis_title='Number of Downloads', yaxis_title='Category')
h_bar.show()
```

```python
cat_no = df_clean.groupby('Category').agg({'App': pd.Series.count})
```

```python
cat_merged_df = pd.merge(cat_no, category_installs, on='Category', how='inner')
cat_merged_df.sort_values('Installs', ascending=False)
```

```python
scatter = px.scatter(cat_merged_df, x='App', y='Installs', title='Category Concentration', size='App', 
                     hover_name=cat_merged_df.index, color='Installs')
scatter.update_layout(xaxis_title='Number of Apps(Lower=More Concentrated)', yaxis_title='Installs', 
                      yaxis=dict(type='log'))
scatter.show()
```

```python
df_fr_vs_pd = df_clean.groupby(['Category', 'Type'], as_index=False).agg({'App': pd.Series.count})
df_fr_vs_pd.head()
```

```python
g_bar = px.bar(df_fr_vs_pd, x='Category', y='App', title='Free vs Paid Apps by Category',
               color='Type', barmode='group')
 
g_bar.update_layout(xaxis_title='Category', yaxis_title='Number of Apps',
                    xaxis={'categoryorder':'total descending'}, yaxis=dict(type='log'))
 
g_bar.show()
```

```python
box = px.box(df_clean, y='Installs', x='Type', color='Type', notched=True, 
             points='all', title='How Many Downloads are Paid Apps Giving Up?')
box.update_layout(yaxis=dict(type='log'))
 
box.show()
```

```python
df_paid_apps = df_clean[df_clean['Type'] == 'Paid']
box = px.box(df_paid_apps, x='Category', y='Revenue_Estimate', title='How Much Can Paid Apps Earn?')
 
box.update_layout(xaxis_title='Category', yaxis_title='Paid App Ballpark Revenue',
                  xaxis={'categoryorder':'min ascending'}, yaxis=dict(type='log'))
 
box.show()
```
